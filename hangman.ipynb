For this week, you will work on a Python Hangman game to practice the Python Data Type skills learned this week. 

```python
import random

words = ["abruptly", "subway", "swivel", "awkward", "transcript", "triphthong", "duplex", "onyx",
    "ivory", "jinx", "kayak", "luxury", "matrix", "oxygen", "pixel", "quiz",
    "sphinx", "vortex", "wizard", "xylophone", "yacht", "zombie", "abyss",
    "alure", "Teddy", "pajama", "oxygen", "ovary", "pneumonia", "jackpot",
    "jigsaw", "kiwifruit", "jukebox", "scratch", "espionage", "strength", "zodiac",
    "bookworm", "zigzag"]

def hangman_game():
    word = random.choice(words)
    word_letters = list(word)  #list of letters
    display = ["_" for letter in word]  # dashes for letters
    guessed = []  #guessed letters
    wrong_guesses = 0  #wrong guess count
    max_wrong = 6  # Max wrong guesses allowed
    
    print("let's playa game of Hangman!!!!")
    print("Word:", " ".join(display))
    
   
    while wrong_guesses < max_wrong and "_" in display: # Loop for playing until 6 wrong guesses or word is guessed
        guess = input("Guess a letter please :)  : ")
        
        
        try:# loop to Check if the input is valid
            if len(guess) != 1:
                print("Please enter just one letter!")
                continue
            if not guess.isalpha():#to check if the input is a valid alphabetic letter
                print("Use Only letters, please!")
                continue
            if guess in guessed:
                print("You already guessed that letter!")
                continue
        except:
            print("Something went wrong. Try again!")
            continue
        
        guessed.append(guess)  # Add guess to guessed letters
        # Check if guess is in the word
        if guess in word_letters:
            print("Good guess!")
        # Replace dashes with the letter
        for i in range(len(word_letters)):
            if word_letters[i] == guess:
                display[i] = guess
        else:
            print("Wrong guess!")
            wrong_guesses = wrong_guesses + 1
        # Show the current state
        print("Word:", " ".join(display))
        print("Guessed letters:", " ".join(guessed))
    # Check if player won or lost
    if "_" not in display:
        print("You won! The word was", word)
    else:
        print("You lost! The word was", word)
hangman_game()
```
