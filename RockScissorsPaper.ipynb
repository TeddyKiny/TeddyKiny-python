```python
import random  #for computer's choice.

while True:  #Main loop 
    print("Enter your Selection \n 1 - Rock \n 2 - Paper \n 3 - Scissors \n")

    #user input
    choice = int(input("Enter your Selection: "))

    #Loop to see if the selection was valid
    while choice > 3 or choice < 1:
        choice = int(input('Please enter a valid Selection:(  : '))

    # Storing a number to a choice
    if choice == 1:
        choice_name = 'Rock'
    elif choice == 2:
        choice_name = 'Paper'
    else:
        choice_name = 'Scissors'

    print('Your choice is:', choice_name)  #So the users Selection is displayed.
    print("I select...")

    # For the computer to select a choice
    comp_choice = random.randint(1, 3)
    if comp_choice == 1:
        comp_choice_name = 'Rock'
    elif comp_choice == 2:
        comp_choice_name = 'Paper'
    else:
        comp_choice_name = 'Scissors'

    print("I selects ", comp_choice_name)  #Display what the computer selected
    print(choice_name, 'vs', comp_choice_name)

    #To check which one should be valid
    if choice == comp_choice:
        result = "DRAW"  # Tie if same.
    elif (choice == 1 and comp_choice == 2) or (comp_choice == 1 and choice == 2):
        result = 'Paper'  # paper beats rock.
    elif (choice == 1 and comp_choice == 3) or (comp_choice == 1 and choice == 3):
        result = 'Rock'  # rock beats scissors.
    elif (choice == 2 and comp_choice == 3) or (comp_choice == 2 and choice == 3):
        result = 'Scissors'  # scissors beats paper.

    # For who won the round
    if result == "DRAW":
        print("<== It's a tie! ==>")
    elif result == choice_name:
        print("<== You wins! ==>")
    else:
        print("<== I'm the winner ==>")

    # To check if the player wants to play again(loop)
    print("Do you want to play again? (Y/N)")
    ans = input().lower()
    if ans == 'n':
        break

print("Hope you had fun") 
```
